# Create Simarlarity Scores Using Cosine

```python
#  Calculate the mean rating for each user
mean_rating = data.groupby(by="user_id", as_index=False)["rating"].mean()

#  Check mean rating for each user
mean_rating

#  merge the two datasets data & mean_rating
data = pd.merge(data, mean_rating, on="user_id", suffixes=(None, "_mean"))

#  Calculate the difference between the rating and the mean rating
data["rating_diff"] = data["rating"] - data["rating_mean"]

# Take a sample of the data
df2 = data.iloc[:data.shape[0]//10]

# Estiamte the size of the memory required
df2.shape[0]**2

# use a pivot table to reshape the data to give us the rating differential on each book
pivot_df = pd.pivot_table(df2, values="rating_diff", index="user_id", columns="book_id")


# Check firs 5 lines of pivot table
pivot_df.head()
```

```python
# Create new dataframe and fill where there is no ratings with mean values
book_df = pivot_df.fillna(pivot_df.mean(axis=0))

# Create new dataframe and fill where there is no ratings with mean values
user_df = pivot_df.apply(lambda row: row.fillna(row.mean()), axis=1)

# Check book df first 5 rows
book_df.head()

# Check user df  first 5 rows
user_df.head()
```

```python
# Calculate the simarlities between users base on their feature vectors
user_simularity_cosine = cosine_similarity(user_df)

# Visualize the vectors scores for each user
user_simularity_cosine

# set the elements of the diagional matrix to o to users are not used againt each other in further analysis
np.fill_diagonal(user_simularity_cosine, 0)

# Creat a new dataframe from the simarility matrix which charts each user against each other
similarity_df = pd.DataFrame(user_simularity_cosine, index=user_df.index, columns=user_df.index)

# Check the first 5 lines of new dataframe
similarity_df.head()
```

```python
# retrieve similarity score for user 1 & 7933
similarity_score_1_7933= user_simularity_cosine[1][10]
print('Similarity score between user 1 and user 933  ', similarity_score_1_7933)
```

```python
# retrieve similarity score for user 1 & 10
similarity_score_1_300= user_simularity_cosine[1][300]
print('Similarity score between user 1 and user 300', similarity_score_1_300)
```

```python
# Pedregosa, F. and Varoquaux, G. and Gramfor (2011). {Scikit-learn: Machine Learning in {P}ython}. https://scikit-learn.org/stable/about.html#citing-scikit-learn. [Accessed 15 May, 2024]
# perform truncated svd on the user feature matrix
number_of_components=12
svd=TruncatedSVD(n_components=number_of_components, random_state=1)
reduced_matrix= svd.fit_transform(user_df)

# Calculate the simarlities between users base on their feature vectors
user_simularity_svd = cosine_similarity(reduced_matrix)

# Visualize the vectors scores for each user
user_simularity_svd

# set the elements of the diagional matrix to o to users are not used againt each other in further analysis
np.fill_diagonal(user_simularity_svd, 0)

# Creat a new dataframe from the simarility matrix which charts each user against each other
similarity_df = pd.DataFrame(user_simularity_svd, index=user_df.index, columns=user_df.index)

# Check the first 5 lines of new dataframe
similarity_df.head()
```

```python
# Find the Eucildean distance between users using linear algebra
def find_distance_between_real_users(userId_1, userId_2):

    ratings_user1=find_users_ratings(userId_1)

    ratings_user2=find_users_ratings(userId_2)

    ratings_comparison=ratings_user1.join(ratings_user2, lsuffix='_user1',
                                           rsuffix='_user2').dropna()

    user1_compared=ratings_comparison['rating_user1']

    user2_compared=ratings_comparison['rating_user2']

    distance_between_users= np.linalg.norm(user1_compared - user2_compared)

    return[userId_1,userId_2, distance_between_users]

result=find_distance_between_real_users(1,3)
print(" Euclidean Distance between users 1 and 3:" , result[2] )
```

```python
# Compute the distance between a specific user 7 and all other users in the dataset
def find_relative_distances(user_id):

    users = ratings_dataframe['user_id'].unique()

    users= users[users != user_id]

    distances= [find_distance_between_real_users(user_id, every_Id) for every_Id in users]

    return pd.DataFrame(distances, columns= ['singleUserId', 'user_id','distances'])

example_distances= find_relative_distances(7)

example_distances


```

```python
# identify users similar to customer 3 based on their distances
def find_top_similar_users(user_id):
    distances_to_user=find_relative_distances(user_id)

    distances_to_user=distances_to_user.sort_values('distances')

    distances_to_user = distances_to_user.set_index('user_id')

    return distances_to_user

Sample_user_ID= 3

find_top_similar_users(Sample_user_ID)
```

```python
#Testing for stationarity of de-trended time series using ADF and KPSS Tests.
#Prathamseh Thakar (2022). Forecasting Time Series Data - Stock Price Analysis. https://medium.com/towards-data-science/forecasting-time-series-data-stock-price-analysis-324bcc520af5. [Accessed 21 May, 2024]
# De-trending the time series
arima['Close_Detrend'] = (arima['Close'] - arima['Close'].shift(30))

def stationarity_test(stock_close_price):
    # Calculating rolling mean and rolling standard deviation:
    rolling_mean = stock_close_price.rolling(30).mean()
    rolling_std_dev = stock_close_price.rolling(30).std()
  
    # Plotting the statistics:
    plt.figure(figsize=(24,6))
    plt.plot(rolling_mean, label='Rolling Mean',linewidth=2.0)
    plt.plot(rolling_std_dev, label = 'Rolling Std Dev',linewidth=2.0)
    plt.plot(stock_close_price,label='De-Trended Time Series')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.tight_layout()
    
    print("ADF Test:")
    adf_test = adfuller(stock_close_price,autolag='AIC')
    print('Null Hypothesis: Not Stationary')
    print('ADF Statistic: %f' % adf_test[0])
    print('p-value: %f' % adf_test[1])
    print('Critical Values:')
    for key, value in adf_test[4].items():
        print('\t%s: %.3f' % (key, value))

    print("KPSS Test:")
    kpss_test = kpss(stock_close_price, regression='c', nlags=None, store=False)
    print('Null Hypothesis: Stationary')
    print('KPSS Statistic: %f' % kpss_test[0])
    print('p-value: %f' % kpss_test[1])
    print('Critical Values:')
    for key, value in kpss_test[3].items():
        print('\t%s: %.3f' % (key, value))
    
stationarity_test(arima['Close_Detrend'].dropna())

# PACF Plot
from statsmodels.graphics.tsaplots import plot_pacf
pacf = plot_pacf(arima['Close_Detrend'].dropna(), lags=30)
```
